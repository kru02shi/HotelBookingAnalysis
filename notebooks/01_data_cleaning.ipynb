```python
import pandas as pd

#Load data
df = pd.read_csv('../data/hotel_bookings.csv')
df.head()
```

```python
df.info()
```

```python
df['children'].value_counts()
```

```python
# Check missing values
df[['children', 'country', 'agent', 'company']].isnull().sum()
```

```python
# Fill missing 'children' with 0
df['children'] = df['children'].fillna(0)
```

```python
df['agent'] = df['agent'].fillna(0)
df['company'] = df['company'].fillna(0)
```

```python
df['country'] = df['country'].fillna('OTH')     #Other country as its not specified
```

```python
# Check missing values
df[['children', 'country', 'agent', 'company']].isnull().sum()
```

```python
#Converting reservation_status_date and arrival date columns to proper datetime format.

df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'])

df['arrival_date'] = pd.to_datetime(df['arrival_date_year'].astype(str) + '-' + df['arrival_date_month'] + '-' + df['arrival_date_day_of_month'].astype(str))
```

```python
df.head()
```

```python
# Encode only select object columns (example)
columns_to_encode = ['hotel', 'meal', 'market_segment', 'distribution_channel', 'deposit_type', 'customer_type']
```

```python
from sklearn.preprocessing import LabelEncoder
enc = LabelEncoder()
for col in columns_to_encode:
    df[col] = enc.fit_transform(df[col])
```

```python
# Saving cleaned version
df.to_csv('../data/hotel_bookings_cleaned.csv', index=False)
```
